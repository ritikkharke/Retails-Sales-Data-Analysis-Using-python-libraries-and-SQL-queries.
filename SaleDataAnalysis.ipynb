```python
import pandas as pd

# CSV file load
df = pd.read_csv("Amazon Sale Report.csv")

# Dataset overview
print("📌 Dataset Shape:", df.shape)
print("\n📌 Column Names:\n", df.columns.tolist())
print("\n📌 Data Types:\n", df.dtypes)
print("\n📌 Missing Values:\n", df.isnull().sum())
print("\n📌 Duplicate Records:", df.duplicated().sum())

# First 5 rows
print("\n📌 First 5 Rows:\n", df.head())

# Numerical columns summary
print("\n📌 Numerical Summary:\n", df.describe())

# Categorical columns summary
print("\n📌 Categorical Summary:\n", df.describe(include='object'))

# Unique values in each column
print("\n📌 Unique Values Count:\n", df.nunique())
```

```python
for col in df.columns:
    if df[col].dtype in ['int64', 'float64']:
        df[col] = df[col].fillna(df[col].median())
    else:
        df[col] = df[col].fillna("Unknown")

```

```python
import pandas as pd

# Step 2 - Remove duplicates
df.drop_duplicates(inplace=True)

# Step 3 - Standardize data types
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
numeric_cols = ['Qty', 'Amount']  # apne dataset ke numeric columns ka naam yaha do
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Step 4 - Fix data inconsistencies (example for 'Category' column)
if 'Category' in df.columns:
    df['Category'] = df['Category'].str.strip().str.title()

# Step 5 - Remove unrealistic values
df = df[(df['Qty'] > 0) & (df['Qty'] < 1000)]
df = df[(df['Amount'] >= 0) & (df['Amount'] < 1000000)]

# Step 6 - Handle outliers using IQR method
for col in ['Qty', 'Amount']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

# Step 7 - Create additional useful columns
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Month_Name'] = df['Order Date'].dt.strftime('%B')

# Step 8 - Save cleaned dataset
df.to_csv("Cleaned_Amazon_Sales.csv", index=False)

print("Data cleaning complete. Cleaned file saved as 'Cleaned_Amazon_Sales.csv'")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# --- 0) Ensure data is loaded & Date is datetime ---
# df = pd.read_csv("Amazon Sale Report.csv")   # <- agar load nahi kiya to uncomment
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df = df.dropna(subset=['Date'])

# --- 1) CustomerID (proxy) bana lo agar real ID nahi hai ---
if 'CustomerID' not in df.columns:
    # Proxy: city + state (best effort). Agar tumhare pass real Customer ID ho to woh use karo.
    df['CustomerID'] = df['ship-city'].astype(str) + "_" + df['ship-state'].astype(str)

# --- 2) Reference date: last transaction + 1 day ---
ref_date = df['Date'].max() + pd.Timedelta(days=1)

# --- 3) Frequency aggregation column decide (Order ID preferred) ---
if 'Order ID' in df.columns:
    freq_col = 'Order ID'
    freq_agg = 'nunique'
else:
    # fallback: count of transactions
    freq_col = 'Date'
    freq_agg = 'count'

# --- 4) RFM base table ---
rfm = df.groupby('CustomerID').agg(
    Recency=('Date', lambda x: (ref_date - x.max()).days),
    Frequency=(freq_col, freq_agg),
    Monetary=('Amount', 'sum')
)

# --- 5) Safe quartiles using ranks (ties handled) ---
def make_quartile(series, reverse=False, max_bins=4):
    # Unique ranking to avoid duplicate bin edges
    ranks = series.rank(method='first')
    q = min(max_bins, max(1, series.nunique()))  # if unique values < 4, reduce bins
    labels = list(range(q, 0, -1)) if reverse else list(range(1, q+1))
    # qcut on ranks avoids duplicate edges; duplicates='drop' is a second safety net
    return pd.qcut(ranks, q, labels=labels, duplicates='drop')

# Recency: lower is better -> reverse True
rfm['R_score'] = make_quartile(rfm['Recency'], reverse=True)
rfm['F_score'] = make_quartile(rfm['Frequency'], reverse=False)
rfm['M_score'] = make_quartile(rfm['Monetary'], reverse=False)

rfm[['R_score','F_score','M_score']] = rfm[['R_score','F_score','M_score']].astype(int)
rfm['RFM_Score'] = rfm['R_score'].astype(str) + rfm['F_score'].astype(str) + rfm['M_score'].astype(str)

# --- 6) Simple segmentation rules ---
def segment(row):
    if row['R_score']>=3 and row['F_score']>=3 and row['M_score']>=3:
        return 'Best Customers'
    if row['F_score']>=3 and row['R_score']>=2:
        return 'Loyal Customers'
    if row['M_score']>=3:
        return 'Big Spenders'
    if row['R_score']>=3:
        return 'Recent Customers'
    return 'At Risk'

rfm['Segment'] = rfm.apply(segment, axis=1)

# --- 7) Summary & quick viz ---
summary = rfm.groupby('Segment').agg(
    Customers=('Segment','count'),
    Avg_Recency=('Recency','mean'),
    Avg_Frequency=('Frequency','mean'),
    Avg_Monetary=('Monetary','mean')
).sort_values('Customers', ascending=False)

print(summary)

seg_counts = rfm['Segment'].value_counts()
plt.figure(figsize=(7,5))
plt.bar(seg_counts.index, seg_counts.values)
plt.title('Customer Segments (RFM)')
plt.xticks(rotation=20)
plt.ylabel('Customers')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ---- Assuming RFM table already calculated ----
# rfm = (Recency, Frequency, Monetary, R_score, F_score, M_score, RFM_Score)

# Step 1: Define segments using RFM scores
def assign_segment(row):
    r, f, m = row['R_score'], row['F_score'], row['M_score']
    
    if r >= 4 and f >= 4 and m >= 4:
        return "High-Value Customers"
    elif f >= 4 and m >= 3:
        return "Loyal Customers"
    elif r <= 2 and f >= 3:
        return "At-Risk Customers"
    elif r >= 4 and f <= 2 and m <= 2:
        return "New Customers"
    elif r <= 2 and f <= 2 and m <= 2:
        return "Lost Customers"
    elif f >= 3 and m >= 3:
        return "Potential Loyalists"
    else:
        return "Others"

rfm['Segment'] = rfm.apply(assign_segment, axis=1)

# Step 2: Segment summary
segment_summary = rfm.groupby('Segment').agg(
    Customers=('Segment','count'),
    Avg_Recency=('Recency','mean'),
    Avg_Frequency=('Frequency','mean'),
    Avg_Monetary=('Monetary','mean')
).sort_values('Customers', ascending=False)

print(segment_summary)

# Step 3: Visualization
plt.figure(figsize=(8,5))
rfm['Segment'].value_counts().plot(kind='bar', color="skyblue")
plt.title("Customer Segments Distribution")
plt.xlabel("Segments")
plt.ylabel("Number of Customers")
plt.xticks(rotation=20)
plt.show()
```

```python
# =========================
# Subtask 2: Forecasting
# =========================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 0) START: ensure df is in memory; else read it
try:
    df
except NameError:
    df = pd.read_csv("Amazon Sale Report.csv", low_memory=False)

# 1) Clean basic fields
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df = df.dropna(subset=['Date'])

# Amount ko numeric banao (agar strings/mixed hue to handle ho jaye)
df['Amount'] = pd.to_numeric(df['Amount'], errors='coerce')
df = df.dropna(subset=['Amount'])

# 2) Daily aggregate -> then Monthly
daily = df.groupby(pd.Grouper(key='Date', freq='D'))['Amount'].sum().asfreq('D', fill_value=0)
monthly = daily.resample('MS').sum()   # Month Start frequency

# 3) Train/Test split (last h months = test)
#   If you have >= 24 months, test horizon 6 months; else ~20%
h = 6 if len(monthly) >= 24 else max(1, int(len(monthly)*0.2))
train = monthly.iloc[:-h]
test  = monthly.iloc[-h:]

print(f"Total months: {len(monthly)}, Train: {len(train)}, Test: {len(test)}, Horizon: {h}")

# 4) Utility metrics
def rmse(y_true, y_pred):
    return float(np.sqrt(np.mean((np.array(y_true) - np.array(y_pred))**2)))

def mape(y_true, y_pred):
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    mask = y_true != 0
    if mask.sum() == 0:
        return np.nan
    return float(np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100)

results = {}

# 5) Model A: Holt-Winters (Exponential Smoothing)
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# seasonal='mul' works well if values are strictly positive; else use 'add'
seasonal_type = 'mul' if (train > 0).all() else 'add'
seasonal_periods = 12 if len(train) >= 24 else None  # only if enough history

try:
    if seasonal_periods:
        hw_model = ExponentialSmoothing(
            train,
            trend='add',
            seasonal=seasonal_type,
            seasonal_periods=seasonal_periods,
            initialization_method='estimated'
        ).fit(optimized=True)
    else:
        # No seasonality fallback
        hw_model = ExponentialSmoothing(
            train,
            trend='add',
            initialization_method='estimated'
        ).fit(optimized=True)

    hw_forecast = hw_model.forecast(h)
    hw_rmse = rmse(test, hw_forecast)
    hw_mape = mape(test, hw_forecast)
    results['Holt-Winters'] = {'rmse': hw_rmse, 'mape': hw_mape, 'forecast': hw_forecast}
except Exception as e:
    print("HW error:", e)

# 6) Model B: SARIMAX (ARIMA with seasonality)
from statsmodels.tsa.statespace.sarimax import SARIMAX

# A safe default seasonal ARIMA; you can tune later
order = (1,1,1)
seasonal_order = (1,1,1,12) if len(train) >= 24 else (0,0,0,0)

try:
    arima_model = SARIMAX(train, order=order, seasonal_order=seasonal_order,
                          enforce_stationarity=False, enforce_invertibility=False)
    arima_fit = arima_model.fit(disp=False)
    arima_forecast = arima_fit.forecast(steps=h)
    arima_rmse = rmse(test, arima_forecast)
    arima_mape = mape(test, arima_forecast)
    results['SARIMAX'] = {'rmse': arima_rmse, 'mape': arima_mape, 'forecast': arima_forecast}
except Exception as e:
    print("SARIMAX error:", e)

# 7) Compare models
res_table = pd.DataFrame(
    {name: {'RMSE': v['rmse'], 'MAPE%': v['mape']} for name, v in results.items()}
).T.sort_values('RMSE')
print("\nModel comparison:")
print(res_table)

best_model_name = res_table.index[0]
best_forecast = results[best_model_name]['forecast']

# 8) Plot backtest (train vs test vs predictions)
plt.figure(figsize=(11,6))
plt.plot(train.index, train.values, label='Train')
plt.plot(test.index, test.values, label='Test', linewidth=3)
plt.plot(test.index, best_forecast.values, label=f'Pred ({best_model_name})', linestyle='--', marker='o')
plt.title(f'Backtest: {best_model_name} | RMSE={results[best_model_name]["rmse"]:.0f} | MAPE={results[best_model_name]["mape"]:.2f}%')
plt.xlabel('Month')
plt.ylabel('Sales Amount')
plt.legend()
plt.tight_layout()
plt.show()

# 9) Fit best model on FULL data and forecast next N months
N = 12  # forecast horizon
if best_model_name == 'Holt-Winters':
    if seasonal_periods:
        final_model = ExponentialSmoothing(
            monthly,
            trend='add',
            seasonal=seasonal_type,
            seasonal_periods=seasonal_periods,
            initialization_method='estimated'
        ).fit(optimized=True)
    else:
        final_model = ExponentialSmoothing(
            monthly,
            trend='add',
            initialization_method='estimated'
        ).fit(optimized=True)
    future_forecast = final_model.forecast(N)

else:  # SARIMAX
    final_arima = SARIMAX(monthly, order=order, seasonal_order=seasonal_order,
                          enforce_stationarity=False, enforce_invertibility=False).fit(disp=False)
    future_forecast = final_arima.forecast(steps=N)

# 10) Show future 12-month forecast
future_df = future_forecast.to_frame(name='forecast')
print("\nNext 12 months forecast:")
print(future_df)

# 11) Plot full series + future
plt.figure(figsize=(11,6))
plt.plot(monthly.index, monthly.values, label='History')
plt.plot(future_df.index, future_df['forecast'].values, label=f'Forecast ({best_model_name})', linestyle='--', marker='o')
plt.title(f'12-Month Sales Forecast ({best_model_name})')
plt.xlabel('Month')
plt.ylabel('Sales Amount')
plt.legend()
plt.tight_layout()
plt.show()

# 12) (Optional) Save outputs
res_table.to_csv('model_comparison_metrics.csv')
future_df.to_csv('sales_forecast_next_12m.csv')
```
